```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.facecolor'] = 'white'  # Figure background
plt.rcParams['axes.facecolor'] = 'white'   # Axes background
plt.rcParams['figure.dpi'] = 100
```

```python
df = pd.read_csv('proj_data.csv', index_col=0)
df['date'] = pd.to_datetime(df['date'])
df
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
plt.subplot(221)
btc = df[df.tick == 'BTC-USD']
plt.plot(btc['date'], btc['adj_prc'])
plt.title('Price of Bitcoin')
plt.tight_layout()  
plt.xlabel('Date')
plt.ylabel('Price ($)')

plt.subplot(222)
eth = df[df.tick == 'ETH-USD']
plt.plot(eth['date'], eth['adj_prc'], color='green')
plt.title('Price of Ethereum')
plt.tight_layout()  
plt.xlabel('Date')
plt.ylabel('Price ($)')

plt.subplot(223)
doge = df[df.tick == 'DOGE-USD']
plt.plot(doge['date'], doge['adj_prc'], color='orange')
plt.title('Price of Dogecoin')
plt.tight_layout()  
plt.xlabel('Date')
plt.ylabel('Price ($)')

plt.subplot(224)
bnb = df[df.tick == 'BNB-USD']
plt.plot(bnb['date'], bnb['adj_prc'], color='lightcoral')
plt.title('Price of Binance Coin')
plt.tight_layout()
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
plt.subplot(221)
plt.hist(btc['logret'], bins=100)
plt.title('Histogram of Historical Bitcoin Log Returns')
plt.tight_layout()  
plt.xlabel('Daily Log Return')
plt.ylabel('Frequency')

plt.subplot(222)
plt.hist(eth['logret'], color = 'green', bins=100)
plt.title('Histogram of Historical Etherum Log Returns')
plt.tight_layout()  
plt.xlabel('Daily Log Return')
plt.ylabel('Frequency')


plt.subplot(223)
plt.hist(doge['logret'], color = 'orange', bins=100)
plt.title('Histogram of Historical Dogecoin Log Returns')
plt.tight_layout()  
plt.xlabel('Daily Log Return')
plt.ylabel('Frequency')

plt.subplot(224)
plt.hist(bnb['logret'], color = 'lightcoral', bins=100)
plt.title('Histogram of Historical Binance Coin Log Returns')
plt.tight_layout()  
plt.xlabel('Daily Log Return')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```
